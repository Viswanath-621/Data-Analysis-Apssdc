```python
days  = [1,2,3,4,5]
temp = [36,35,37.7,40,38.2]
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=10)
plt.plot(days,temp,c ='g',marker='o',linewidth=2)

plt.show()
```

```python
v_temp = [40.5,43.4,44.6,39.2,38.9]
plt.plot(days,v_temp,'m*--',linewidth=2,label='vij')
plt.plot(days,temp,c ='g',marker='o',linewidth=2)
plt.title('gnt-temp vs vij_temp')
plt.xlabel('days')
plt.ylabel('temperature')

plt.show()
```

```python
v_temp = [40.5,43.4,44.6,39.2,38.9]
plt.plot(days,v_temp,'m*--',linewidth=2,label='vij')
plt.plot(days,temp,c ='g',marker='o',linewidth=2)
plt.title('gnt-temp vs vij_temp')
plt.xlabel('days')
plt.ylabel('temperature')
plt.legend(loc=4)
plt.grid(c ='r',linestyle='-',linewidth=2)
plt.show()
```
